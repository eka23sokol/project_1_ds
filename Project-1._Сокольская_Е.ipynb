1. Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:
* Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

*Совет: проверяйте полученные категории, например, с помощью метода unique()*


```python
def get_education(education):
    list_education = education.split(' ')
    level_education = list_education[:2]
    if level_education[0] == 'Высшее':
        return 'высшее'
    if level_education[0] == 'Неоконченное':
        return 'неоконченное высшее'
    if level_education[0] == 'Среднее' and level_education[1] == 'специальное':
        return 'среднее специальное'
    else:
        return 'среднее'

hh_data['Образование'] = hh_data['Образование и ВУЗ'].apply(get_education)

hh_data = hh_data.drop(['Образование и ВУЗ'], axis=1)
```

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**.

Для начала обсудим условия решения задачи:

* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def get_experience(experience):
    if experience is np.nan or experience == 'Не указано':
        return np.nan
    year_words = ['год', 'года', 'лет']
    month_words = ['месяц', 'месяцев', 'месяца']
    list_exp = experience.split()[2:7]
    years = 0
    months = 0
    for index, item in enumerate(list_exp):
        if item in year_words:
            years = int(list_exp[index-1])
        if item in month_words:
            months = int(list_exp[index-1])
    return int(years*12 + months)

#get_experience('Опыт работы 10 лет 11 месяцев  Сентябрь 2017 — по настоящее время 1 год 9 месяцев TNT Express Worldwide (CIS), Moscow offic')

hh_data['Опыт работы (месяц)'] = hh_data['Опыт работы'].apply(get_experience)

hh_data = hh_data.drop(['Опыт работы'], axis=1)
```

Создадим отдельные признаки **"Город"**, **"Готовность к переезду"**, **"Готовность к командировкам"**. При этом важно учесть:

    * … , готов к командировкам , …

    Еще один важный факт: при выгрузки данных у некоторых соискателей "потерялась" информация о готовности к командировкам. Давайте по умолчанию будем считать, что такие соискатели не готовы к командировкам.
    
Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
def get_city(city):
    million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград' ]
    capital_city = ['Москва', 'Санкт-Петербург']
    city_split = city.split(' , ')[:1]
    if city_split[0] in capital_city:
        return city_split[0]
    elif city_split[0] in million_cities:
        return 'город-миллионник'
    else:
        return 'другие'
    
def get_moving(moving):
    moving_split = moving.split(' , ')
    mov_words = ['переезду', 'переехать']
    for index, item in enumerate(moving_split):
        for word in item.split():
            if word in mov_words:
                for i in moving_split[index].split():
                    if i == 'не':
                        return False
                else:
                    return True
        
def get_trip(trip):
    trip_split = trip.split(' , ')
    for index, item in enumerate(trip_split):
        for word in item.split():
            if word == 'командировкам':
                for i in trip_split[index].split():
                    if i == 'не':
                        return False
                else:
                    return True
    return False


hh_data['Город'] = hh_data['Город, переезд, командировки'].apply(get_city)
hh_data['Готовность к переезду'] = hh_data['Город, переезд, командировки'].apply(get_moving)
hh_data['Готовность к командировкам'] = hh_data['Город, переезд, командировки'].apply(get_trip)
hh_data = hh_data.drop(['Город, переезд, командировки'], axis=1)
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
def get_full_employment(employment):
    employment_split = employment.split(', ')
    for item in employment_split:
        if item == 'полная занятость':
            return True
        continue
    return False
    
def get_part_employment(employment):
    employment_split = employment.split(', ')
    for item in employment_split:
        if item == 'частичная занятость':
            return True
        continue
    return False

def get_project_work(employment):
    employment_split = employment.split(', ')
    for item in employment_split:
        if item == 'проектная работа':
            return True
        continue
    return False

def get_volunteering(employment):
    employment_split = employment.split(', ')
    for item in employment_split:
        if item == 'волонтерство':
            return True
        continue
    return False

def get_internship(employment):
    employment_split = employment.split(', ')
    for item in employment_split:
        if item == 'стажировка':
            return True
        continue
    return False

hh_data['полная занятость'] = hh_data['Занятость'].apply(get_full_employment)
hh_data['частичная занятость'] = hh_data['Занятость'].apply(get_part_employment)
hh_data['проектная работа'] = hh_data['Занятость'].apply(get_project_work)
hh_data['волонтерство'] = hh_data['Занятость'].apply(get_volunteering)
hh_data['стажировка'] = hh_data['Занятость'].apply(get_internship)

def get_full_day(schedule):
    schedule_split = schedule.split(', ')
    for item in schedule_split:
        if item == 'полный день':
            return True
        continue
    return False

def get_shift_work(schedule):
    schedule_split = schedule.split(', ')
    for item in schedule_split:
        if item == 'сменный график':
            return True
        continue
    return False

def get_flexible_schedule(schedule):
    schedule_split = schedule.split(', ')
    for item in schedule_split:
        if item == 'гибкий график':
            return True
        continue
    return False

def get_distant_work(schedule):
    schedule_split = schedule.split(', ')
    for item in schedule_split:
        if item == 'удаленная работа':
            return True
        continue
    return False

def get_watch(schedule):
    schedule_split = schedule.split(', ')
    for item in schedule_split:
        if item == 'вахтовый метод':
            return True
        continue
    return False

hh_data['полный день'] = hh_data['График'].apply(get_full_day)
hh_data['сменный график'] = hh_data['График'].apply(get_shift_work)
hh_data['гибкий график'] = hh_data['График'].apply(get_flexible_schedule)
hh_data['удаленная работа'] = hh_data['График'].apply(get_distant_work)
hh_data['вахтовый метод'] = hh_data['График'].apply(get_watch)

hh_data = hh_data.drop(['Занятость', 'График'], axis=1)
```

4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
# выгружаем курсы валют, которые встречаются в наших данных за период с 29.12.2017 по 05.12.2019.
Exchange_data = pd.read_csv('data/ExchangeRates.csv')

# обновляем признак Обновление резюме и выделяем дату
hh_data['Обновление резюме'] = pd.to_datetime(hh_data['Обновление резюме']).dt.date


# обновляем признак date и выделяем дату
Exchange_data['date'] = pd.to_datetime(Exchange_data['date']).dt.date

# функция преобразования наименования валюты в стандарт ISO
def get_salary_сurrency(salary):
    currency_dict = {
        'USD': 'USD', 'KZT': 'KZT',
        'грн': 'UAH', 'белруб': 'BYN',
        'EUR': 'EUR', 'KGS': 'KGS',
        'сум': 'UZS', 'AZN': 'AZN'
    }
    currency = salary.split(' ')[1].replace('.','')
    if currency == "руб":
        return "RUB"
    else:
        return currency_dict[currency]
    

hh_data['ЗП (tmp)'] = hh_data['ЗП'].apply(lambda x: float(x.split(' ')[0]))
hh_data['Валюта'] = hh_data['ЗП'].apply(get_salary_сurrency)


# объединяем таблицы, применяем тип объединения left
merged_data = hh_data.merge(
    Exchange_data,
    left_on= ['Валюта', 'Обновление резюме'],
    right_on= ['currency', 'date'],
    how = 'left'
)
#заполняем пропуски
merged_data['close'] =  merged_data['close'].fillna(value=1)
merged_data['proportion'] = merged_data['proportion'].fillna(value=1)


hh_data['ЗП (руб)'] = merged_data['ЗП (tmp)'] * merged_data['close'] / merged_data['proportion']
hh_data = hh_data.drop(['ЗП', 'ЗП (tmp)', 'Валюта'], axis=1)


#print(round(hh_data['ЗП (руб)'].median()/1000))
```

Из графика видим, что среднее значение проходит правее моды, следовательно наблюдается правосторонняя асимметрия.

```python
def outliers_z_score_mod(data, feature, log_scale=False, left=3, right=3):
    if log_scale:
        x = np.log(data[feature]+1)
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned

outliers, cleaned = outliers_z_score_mod(hh_data, 'Возраст', log_scale=True, left=3, right=4)
print(f'Число выбросов по методу z-отклонения: {outliers.shape[0]}')
print(f'Результирующее число записей: {cleaned.shape[0]}')
```
